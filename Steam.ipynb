# Top 10 com a melhor review

Percentagem das Reviews

```python

vals = []
for s in data.recent_reviews:
    valor = 0
    try:
        #print(s)
        index = s.find('%')
       
    
        a=s[index -2]
        b=s[index -1]
        valor=str(a)+str(b)
    except:
        valor = 0
        
    #print(valor)
    vals.append(valor)
    
```

Cria uma nova coluna com os valores

```python
data['Score'] = vals
data.head()
```

Muda de objeto para float

```python
data['Score'] = data["Score"].astype(float)

#display data types
print(data.dtypes)
```

Os 10 jogos com a melhor review

```python
data.nlargest(n=10, columns=['Score'])
```

Tabela Simples

```python
D = developerindex['Developer'].head(10)
Q = developerindex ['Quantidade'].head(10)

fig = plt.figure(figsize =(10, 7))
plt.bar(D[0:10], Q[0:10])
plt.xticks(rotation='vertical')
plt.show()
```

Tabela mais complexa

```python
D = developerindex['Developer'].head(10)
Q = developerindex ['Quantidade'].head(10)

fig, ax = plt.subplots(figsize =(16, 9))
 
ax.barh(D, Q)
 

for s in ['top', 'bottom', 'left', 'right']:
    ax.spines[s].set_visible(False)
 
ax.xaxis.set_ticks_position('none')
ax.yaxis.set_ticks_position('none')
 
ax.xaxis.set_tick_params(pad = 5)
ax.yaxis.set_tick_params(pad = 10)
 

ax.grid(b = True, color ='grey',
        linestyle ='-.', linewidth = 0.5,
        alpha = 0.2)
 
ax.invert_yaxis()
 
for i in ax.patches:
    plt.text(i.get_width()+0.2, i.get_y()+0.5,
             str(round((i.get_width()), 2)),
             fontsize = 10, fontweight ='bold',
             color ='grey')
 
ax.set_title('Top Dez Developers com Mais Jogos',
             loc ='left', )

plt.show()
```

# Top 10 melhores descontos

Tirar os simbolos

```python
data['original_price'] = data['original_price'].str.replace('$','', regex=True)
data['discount_price'] = data['discount_price'].str.replace('$','', regex=True)

data.head()
```

Por o 'Free' e 'nan' como '0'
Troca todas as frases restantes para numerico
Troca todos NaN para '0'

```python
data['original_price'].replace({"Free": "0"}, inplace=True)
data['original_price'].replace({"Free To Play": "0"}, inplace=True)
data['original_price'].replace({"nan": "0"}, inplace=True)

data['original_price'] = pd.to_numeric(data['original_price'],errors='coerce')
data = data.replace(np.nan, 0, regex=True)


data['original_price'] = data['original_price'].fillna(0)
data['discount_price'] = data['discount_price'].fillna(0)

data.head()
```

Passa a coluna de objeto para float

```python
data['original_price'] = pd.to_numeric(data['original_price'], downcast='float')
data['discount_price'] = pd.to_numeric(data['discount_price'], downcast='float')

data.dtypes
```

Cria coluna com a diferença de preço

```python
data['diferenca'] = data.apply(lambda x: x['original_price'] - x['discount_price'], axis=1)

data.head(3)
```

Calcular o desconto

```python
data['desc'] = data['diferenca'] / data['original_price']
data['descontoTotal'] = data['desc'] * 100
data.head()
```

Retirar os 100 em desconto total

```python
f = data['descontoTotal']
data.loc[data['descontoTotal'] == 100, 'descontoTotal'] = 0
f.head()
```

Top 10 Melhores Descontos

```python
data.nlargest(n=10, columns=['descontoTotal'])
```
